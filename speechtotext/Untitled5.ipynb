```python
# Import libraries
import whisper
from pydub import AudioSegment
from pydub.silence import split_on_silence
```

```python
from google.colab import files  #elhsah
uploaded = files.upload()   # Upload the audio file here

# Define the process_audio function
import subprocess

def process_audio(file_name: str, task: str = "transcribe", model: str = "medium") -> str:
    try:
        # Run the whisper command with the specified arguments
        result = subprocess.run(
            ["whisper", file_name, "--model", model, "--task", task],
            capture_output=True,
            text=True
        )
        # Check for errors
        if result.returncode != 0:
            raise Exception(f"Error in Whisper: {result.stderr}")

        return result.stdout
    except Exception as e:
        return f"An error occurred: {str(e)}"

file_name = list(uploaded.keys())[0]  # Get the uploaded file name
output = process_audio(file_name, task="translate", model="medium")
print(output)
```
